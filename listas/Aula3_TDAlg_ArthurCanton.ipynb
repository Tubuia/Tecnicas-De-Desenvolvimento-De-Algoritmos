```python
'''
Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
números a serem lidos. Em seguida, leia n números (conforme o valor informado
anteriormente) e imprima a soma e a média aritmética dos números informados.
'''
print("--Exercicio 11--")
n = int(input("Digite a quantidade de números: "))
soma = 0
for i in range(n):
    numero = float(input(f"Digite o {i+1}º número: "))
    soma += numero
media = soma / n
print(f"Soma dos números: {soma}")
print(f"Média aritmética: {media}")
```

```python
'''
Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o nome e idade de N
pessoas e ao final apresentar o nome da pessoa mais velha.
'''
print("--Exercicio 12--")
n = int(input("Digite a quantidade de números: "))
nome_mais_velha = ""
idade_mais_velha = 0
for i in range(n):
  print(f"Pessoa {i + 1}:")
  nome = input("Digite o nome: ")
  idade = int(input("Digite a idade: "))
  if idade > idade_mais_velha:
    idade_mais_velha = idade
    nome_mais_velha = nome
print(f"A pessoa mais velha é {nome_mais_velha} com {idade_mais_velha} anos")
```

```python
'''
Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.
'''
print("--Exercicio 13--")
n = int(input("Digite a quantidade de pessoas: "))
soma_idade_m = 0
soma_idade_f = 0
quantidade_m = 0
quantidade_f = 0
for i in range(n):
    print(f"Pessoa {i+1}:")
    while True:
        sexo = input("Digite o sexo (M/F): ").strip().upper()
        if sexo in ['M', 'F']:
            break
        else:
            print("Sexo inválido. Digite 'M' para masculino ou 'F' para feminino.")
    while True:
        try:
          idade = int(input("Digite a idade: "))
          break
        except ValueError:
            print("Por favor, digite um número válido para a idade.")
    if sexo == 'M':
        soma_idade_m += idade
        quantidade_m += 1
    else:
        soma_idade_f += idade
        quantidade_f += 1
media_m = soma_idade_m / quantidade_m if quantidade_m > 0 else 0
media_f = soma_idade_f / quantidade_f if quantidade_f > 0 else 0
print(f"Média de idade dos homens: {media_m:.2f}")
print(f"Média de idade das mulheres: {media_f:.2f}")
```

```python
'''
Elabore um algoritmo que solicite ao usuário 10 números reais e ao final apresente o
maior e o menor deles.
'''
print("--Exercicio 14--")
numero = float(input("Digite o 1º número real: "))
maior = numero
menor = numero
for i in range(2, 11):
    numero = float(input(f"Digite o {i}º número real: "))
    if numero > maior:
        maior = numero
    if numero < menor:
        menor = numero
print(f"Maior número: {maior}")
print(f"Menor número: {menor}")
```

```python
'''
Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.
'''
print("--Exercicio 15--")
numeros = []
print("Digite números reais (digite 0 para encerrar):")
while True:
    try:
        numero = float(input("Digite um número: "))
        if numero == 0:
            break
        numeros.append(numero)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número real válido.")
numeros.sort()
for num in numeros:
    print(num)
```

```python
'''
Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.
'''
print("--Exercicio 16--")
def calcular_mediana(lista):
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)
    meio = n // 2

    if n % 2 == 0:
        return (lista_ordenada[meio - 1] + lista_ordenada[meio]) / 2
    else:
        return lista_ordenada[meio]

def main():
    idades = []

    while True:
        try:
            idade = int(input("Informe a idade do aluno: "))
            idades.append(idade)
        except ValueError:
            print("Por favor, insira um número válido.")
            continue

        continuar = input("Deseja informar a idade do próximo aluno? (s/n): ").strip().lower()
        if continuar != 's':
            break

    if not idades:
        print("Nenhuma idade foi informada.")
        return

    mais_novo = min(idades)
    mais_velho = max(idades)
    acima_18 = sum(1 for i in idades if i > 18)
    ate_18 = sum(1 for i in idades if i <= 18)
    media = sum(idades) / len(idades)
    mediana = calcular_mediana(idades)

    print(f"Idade do mais novo: {mais_novo}")
    print(f"Idade do mais velho: {mais_velho}")
    print(f"Quantidade de alunos com mais de 18 anos: {acima_18}")
    print(f"Quantidade de alunos com até 18 anos: {ate_18}")
    print(f"Média das idades: {media:.2f}")
    print(f"Mediana das idades: {mediana:.2f}")

if __name__ == "__main__":
    main()
```
